1) ¿Cuál es el tamaño en bytes de un array de 1000
elementos de tipo booleano?

4) Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9],
obtenga la media aritmética de la diferencia (a-b).

```python
import numpy as np

a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

diferencia = a - b
media_diferencia = np.mean(diferencia)

print(f"La media aritmética de la diferencia (a-b) es: {media_diferencia}")
```

## Parte 2 Exploración y minería de datos

Con el dataset nucleospoblacion.csv adjunto al laboratorio, luego responda las siguientes preguntas usando pandas y
matplotlib.

1) ¿Cuántos Municipios tienen más de 100000 habitantes?.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import euclidean
```

```python
# Cargar el conjunto de datos
datos = pd.read_csv('nucleospoblacion.csv')
```

```python
# 1) ¿Cuántos Municipios tienen más de 100,000 habitantes?
municipios_mas_100k = datos[datos['Poblacion'] > 100000]
cantidad_municipios_mas_100k = len(municipios_mas_100k)
print(f"1) La cantidad de municipios con más de 100,000 habitantes es: {cantidad_municipios_mas_100k}")
```

```python
# 2) Realice una gráfica de barras sobre la población de cada ciudad
#    a. ¿Cuál es la segunda ciudad más poblada?
#    b. ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?

# Ordenar por población
datos_sorted = datos.sort_values(by='Poblacion')

# a. Segunda ciudad más poblada
segunda_ciudad = datos_sorted['Municipio'].iloc[-2]
print(f"2a) La segunda ciudad más poblada es: {segunda_ciudad}")

# b. Posición de Granada en el ranking
posicion_granada = datos_sorted[datos_sorted['Municipio'] == 'Granada'].index[0] + 1
print(f"2b) Granada ocupa la posición {posicion_granada} en el ranking de las ciudades más pobladas.")

# Graficar
plt.figure(figsize=(12, 6))
plt.bar(datos_sorted['Municipio'], datos_sorted['Poblacion'])
plt.xticks(rotation=90)
plt.xlabel('Municipio')
plt.ylabel('Población')
plt.title('Población de cada municipio')
plt.show()
```

```python
# 3) ¿Cuántos municipios de Extremadura tienen más de 5000 habitantes?
municipios_extremadura = datos[(datos['Provincia'] == 'Badajoz') | (datos['Provincia'] == 'Cáceres')]
cantidad_municipios_extremadura_mas_5000 = len(municipios_extremadura[municipios_extremadura['Poblacion'] > 5000])
print(f"3) La cantidad de municipios de Extremadura con más de 5000 habitantes es: {cantidad_municipios_extremadura_mas_5000}")
```

```python
# 4) ¿Cuál es el municipio situado más al Norte?
municipio_mas_al_norte = datos.loc[datos['Y'].idxmax()]
print(f"4) El municipio más al norte es {municipio_mas_al_norte['Municipio']} en la provincia de {municipio_mas_al_norte['Provincia']} con una población de {municipio_mas_al_norte['Poblacion']} habitantes.")
```

```python
# 5) Obtenga la media, mediana, desviación estándar, valor
#    máximo y valor mínimo de la población de los municipios
#    de la provincia de Granada.
datos_granada = datos[datos['Provincia'] == 'Granada']
media = datos_granada['Poblacion'].mean()
mediana = datos_granada['Poblacion'].median()
desviacion_estandar = datos_granada['Poblacion'].std()
maximo = datos_granada['Poblacion'].max()
minimo = datos_granada['Poblacion'].min()

print(f"5) Estadísticas de la población en la provincia de Granada:")
print(f"   Media: {media}, Mediana: {mediana}, Desviación Estándar: {desviacion_estandar}, Máximo: {maximo}, Mínimo: {minimo}")
```

```python
# 6) Realice un histograma con la población de los
#    Municipios para cada una de las provincias.
plt.figure(figsize=(12, 6))
datos.groupby('Provincia')['Poblacion'].plot(kind='hist', alpha=0.5, legend=True)
plt.xlabel('Población')
plt.ylabel('Frecuencia')
plt.title('Histograma de la población por provincia')
plt.show()
```

```python
# 7) Seleccione al azar cincuenta municipios diferentes de
#    entre los disponibles en el archivo. Asegúrese de que no
#    se repitan. ¿Luego calcule el promedio de la población y
#    la desviación estándar de esto 50 municipios?.
municipios_aleatorios = datos.sample(n=50, random_state=42)
promedio_aleatorio = municipios_aleatorios['Poblacion'].mean()
desviacion_estandar_aleatoria = municipios_aleatorios['Poblacion'].std()

print(f"7) Promedio de población de 50 municipios aleatorios: {promedio_aleatorio}")
print(f"   Desviación estándar de población de 50 municipios aleatorios: {desviacion_estandar_aleatoria}")
```

```python
# 8) Determine los nombres de los Municipios más cercano y
#    más lejano a Madrid. Para ello debe calcular la
#    distancia en todos ellos y Madrid. 
#    No considere a Madrid en el análisis ya que la distancia sería cero.

# Filtrar Madrid
madrid_coord = datos[datos['Municipio'] == 'Madrid'][['X', 'Y']].values[0]
datos_sin_madrid = datos[datos['Municipio'] != 'Madrid'].copy()  # Usar .copy() para evitar la advertencia

# Calcular distancias
datos_sin_madrid['Distancia_Madrid'] = datos_sin_madrid.apply(lambda row: euclidean(madrid_coord, [row['X'], row['Y']]), axis=1)

# Municipio más cercano
municipio_mas_cercano = datos_sin_madrid.loc[datos_sin_madrid['Distancia_Madrid'].idxmin()]

# Municipio más lejano
municipio_mas_lejano = datos_sin_madrid.loc[datos_sin_madrid['Distancia_Madrid'].idxmax()]

print(f"8) El municipio más cercano a Madrid (sin considerar Madrid) es {municipio_mas_cercano['Municipio']} en la provincia de {municipio_mas_cercano['Provincia']} con una distancia de {municipio_mas_cercano['Distancia_Madrid']} unidades.")
print(f"   El municipio más lejano a Madrid (sin considerar Madrid) es {municipio_mas_lejano['Municipio']} en la provincia de {municipio_mas_lejano['Provincia']} con una distancia de {municipio_mas_lejano['Distancia_Madrid']} unidades.")
```
